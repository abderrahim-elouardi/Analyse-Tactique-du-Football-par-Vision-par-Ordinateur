# **Extraction des frames des videos**

```python
output_folder = r"./../data/extracted_frames"
os.makedirs(output_folder, exist_ok=True)
```

```python

video_path = r"./../data/raw_video/FULL_MATCH_ManCity_v_Tottenham.webm"

# Récupérer la durée de la vidéo
probe = ffmpeg.probe(video_path)
duration = float(probe['format']['duration'])

# Extraire une frame toutes les 10 secondes
for t in range(0, int(duration), 5):
    out_file = os.path.join(output_folder, f"frame_{t}.jpg")
    (
        ffmpeg
        .input(video_path, ss=t)
        .output(out_file, vframes=1)
        .run(quiet=True, overwrite_output=True)
    )

print("Extraction terminée !")
```
